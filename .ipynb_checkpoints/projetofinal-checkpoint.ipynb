### Extração dos Dados


```python
import pandas as pd
from google.cloud import bigquery
from google.oauth2 import service_account
from pandas.io import gbq
```

```python
contratos = pd.read_csv ('tabela_contratos.csv')
```

```python
contratos
```

```python
datas = pd.read_csv ('tabela_datas.csv')
```

```python
datas
```

```python
empresas = pd.read_csv ('tabela_empresas.csv')
```

```python
empresas
```

### Transformação dos Dados

```python
contratos_empresas = contratos.merge(empresas,
                                     left_on = 'fk_empresa_contratada',
                                     right_on = 'id_empresa',
                                     how = 'left')
```

```python
contratos_empresas
```

```python
contratos_empresas_datas = contratos_empresas.merge(datas,
                                     left_on = 'inicio_vigencia',
                                     right_on = 'id_data',
                                     how = 'left')
```

```python
contratos_empresas_datas
```

```python
contratos_empresas_datas.drop(columns=['id_data','fk_empresa_contratada','inicio_vigencia','id_empresa'], inplace =True)
```

```python
contratos_empresas_datas
```

```python
contratos_empresas_final = contratos_empresas_datas.merge(datas,
                                     left_on = 'termino_vigencia',
                                     right_on = 'id_data',
                                     how = 'left')
```

```python
contratos_empresas_final
```

```python
contratos_empresas_final.drop(columns=['termino_vigencia','id_data'], inplace=True)
```

```python
contratos_empresas_final
```

```python
contratos_empresas_final.rename(columns={'data_x': 'data_inicio_vigencia'},inplace=True)
```

```python
contratos_empresas_final.rename(columns={'data_y': 'data_termino_vigencia'},inplace=True)
```

```python
contratos_empresas_final
```

### Validando os Valores do DataFrame Contratos_empresas_final

```python
contratos_empresas_final.count()
```

```python
contratos_empresas_final.dtypes
```

```python
contratos_empresas_final.data_inicio_vigencia = pd.to_datetime(contratos_empresas_final.data_inicio_vigencia,
                                                             format='%d/%m/%Y').dt.date
```

```python
contratos_empresas_final.data_termino_vigencia = pd.to_datetime(contratos_empresas_final.data_termino_vigencia,
                                                             format='%d/%m/%Y').dt.date
```

```python
for i in contratos_empresas_final.data_termino_vigencia:
    print (i)
    print (pd.to_datetime(i))
```

```python
contratos_empresas_final.data_termino_vigencia = contratos_empresas_final.data_termino_vigencia.str.replace('31/09/2017','30/09/2017')
```

```python
contratos_empresas_final.data_termino_vigencia = pd.to_datetime(contratos_empresas_final.data_termino_vigencia,
                                                             format='%d/%m/%Y').dt.date
```

```python
contratos_empresas_final.dtypes
```

```python
contratos_empresas_final
```

```python
contratos_empresas_final['tempo_contrato'] = (contratos_empresas_final.data_termino_vigencia - \
                                              contratos_empresas_final.data_inicio_vigencia).dt.days
```

```python
contratos_empresas_final
```

### Validando os valores dos dias de contrato

```python
contratos_empresas_final.tempo_contrato.value_counts()
```

```python
contratos_empresas_final = contratos_empresas_final[contratos_empresas_final.tempo_contrato > 0]
```

```python
contratos_empresas_final.reset_index(drop=True, inplace=True)
```

```python
contratos_empresas_final
```

### Carregamento dos Dados

```python
credentials = service_account.Credentials.from_service_account_file("GBQ.json",
                                                                   scopes=["https://www.googleapis.com/auth/cloud-platform",
                                                                           'https://www.googleapis.com/auth/drive'])
```

```python
contratos_empresas_final.to_gbq(credentials=credentials,
                                destination_table='etlfinal.etl_csv',
                                if_exists='replace',
                                table_schema=[{'name': 'data_inicio_vigencia', 'type': 'DATE'},
                                             {'name': 'data_termino_vigencia', 'type': 'DATE'}])
```
